```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

def show_data(filename):
    df1 = pd.read_csv(filename, names=('T', 'To', 'Sol', 'Tso', 'Tsi', 'Tr'), header=0)

    plt.plot(df1['T'], df1['To'], label="To")
    plt.plot(df1['T'], df1['Tso'], label="Tso")
    plt.plot(df1['T'], df1['Tsi'], label="Tsi")
    plt.plot(df1['T'], df1['Tr'], label="Tr")

    plt.ylim(0,50)
    plt.grid()
    plt.legend(loc="lower right")
    plt.show()
```

* 壁体が熱容量を持っていないため計算の立ち上がりが素早い。
* 熱の応答が速い。

```python
show_data('test.csv')
```

* 非定常計算では、壁体による熱移動の遅延が見て取れる。
* 計算の立ち上がりにも時間がかかる

```python
show_data('test2.csv')
```
